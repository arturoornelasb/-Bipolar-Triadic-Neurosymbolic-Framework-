## 2. Methodology: Triadic Semantic Logic

We solve analogies using **Ratio Logic**. 
The analogy $A : B :: C : D$ implies $D = \frac{C \cdot B}{A}$.

For the classic example: $King : Man :: Queen : Woman$
We compute: $Queen_{calc} = \frac{King \cdot Woman}{Man}$

**Paper Section:** Methodology / Triadic Semantic Logic

```python
from triadic_framework.core.semantic_engine import TriadicRelationalFramework

engine = TriadicRelationalFramework()

val_king = mapper.get_concept_value("king")
val_man = mapper.get_concept_value("man")
val_woman = mapper.get_concept_value("woman")
val_queen_expected = mapper.get_concept_value("queen")

print(f"Calculating: (King * Woman) / Man")
print(f"({val_king} * {val_woman}) / {val_man}")

# Run Analogy
C4, K, steps = engine.analogy_variant(val_king, val_man, val_woman)

print(f"\nResult C4: {C4}")
print(f"Expected Queen: {val_queen_expected}")
print(f"Simplicity K: {K}")

if C4 == val_queen_expected:
    print("\n‚úÖ SUCCESS: The engine correctly derived 'Queen'.")
else:
    print("\n‚ùå FAILURE.")
```

## 8. Application: Semantic Compression

This section demonstrates a practical application of the Semantic Engine: **Semantic Compression**. By mapping words to unique IDs in the vector space, we can achieve compression ratios superior to standard text encoding.

```python
from semantic_compressor import SemanticCompressor
import os

# Create dummy input file
input_text = "hombre rey mujer reina " * 100
with open("test_input.txt", "w") as f:
    f.write(input_text)

print("--- Semantic Compression Test ---")
compressor = SemanticCompressor()

# Compress
compressor.compress("test_input.txt", "test_output.semz")

# Decompress
compressor.decompress("test_output.semz", "restored.txt")

# Verify
with open("restored.txt", "r") as f:
    restored_text = f.read()
    
if restored_text == input_text:
    print("‚úÖ Compression/Decompression Cycle Successful.")
    
    orig_size = os.path.getsize("test_input.txt")
    comp_size = os.path.getsize("test_output.semz")
    ratio = (1 - comp_size/orig_size) * 100
    print(f"Original Size: {orig_size} bytes")
    print(f"Compressed Size: {comp_size} bytes")
    print(f"Compression Ratio: {ratio:.2f}%")
else:
    print("‚ùå Data Corruption Detected.")
```
